# Data Cleaning


```python
data = pd.read_csv('/content/bank-additional-full.csv', delimiter=';', quotechar='"')
```

```python
df=pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
for col in df.columns:
  if df[col].dtype == 'object':
    df[col]=df[col].astype('category')
```

```python
df.describe()
```

```python
df_yes = df[df['y'] == 'yes']
df_no = df[df['y'] == 'no']
for col in df.select_dtypes(include='category').columns:
  plt.figure(figsize=(12, 3))

  plt.subplot(1, 2, 1)
  sns.countplot(x=col, data=df_yes)
  plt.title('Success')
  plt.xticks(rotation=45)

  plt.subplot(1, 2, 2)
  sns.countplot(x=col, data=df_no)
  plt.title('Not Success')
  plt.xticks(rotation=45)

  plt.show()

'''
  job : the top 3 [admin , blue-collar , techinician]
  education : the top is -> university degree
  housing : the majority is YES
  loan : majority is NO
  contact : Cellular is higher
  '''
```

```python
for col in  df.select_dtypes(include=['int64','float64']).columns:
    plt.figure(figsize=(12, 3))

    plt.subplot(1, 2, 1)
    df[col].hist(bins=30)
    plt.title(col)

    plt.subplot(1, 2, 2)
    df[col].plot(kind='box')

    plt.show()
```
